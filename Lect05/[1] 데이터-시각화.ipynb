```python
# 코드 실행에 필요한 패키지 불러오기
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
# matplotlib 로 그림 그리기 : plot (데이터 포인트를 선으로 이은 그래프)
x = [1,2,3,4,5]  # x 축 값
y = [4,6,8,8,11]  # y 축 값
plt.plot(x,y)  # 그래프로 그리기 (아직 화면에 출력하지 않음)
plt.show()  # 그래프를 화면에 보여주기
```

```python
# matplotlib 로 그림 그리기 : scatter (산점도 그래프)
x = [1,2,3,4,5]
y = [4,6,8,8,11]
plt.scatter(x,y)
plt.show()
```

```python
# matplotlib 로 그림 그리기 : histogram/히스토그램 (발생 빈도 수)
x = [1,2,3,4,5]
y = [4,6,8,8,11]
plt.hist(y)
plt.show()
```

```python
# matplotlib 로 그림 그리기 : plot (두개 이상의 데이터)
# x 축 값은 동일하고, y 축 값이 서로 다른 y1과 y2가 있을때...
x = [1,2,3,4,5]  
y1 = [4,6,8,8,11] 
y2 = [2,1,1,5,3]
plt.plot(x,y1,x,y2)
plt.xlabel('x values')  # x 축 레이블 설정
plt.ylabel('y values')  # y 축 레이블 설정
plt.title('Title')  # 그래프 제목 설정하기
plt.legend(['y1', 'y2'])  # 범례 설정하기
plt.show()
```

```python
# 서울 기온 데이터(csv) 불러오기
# CSV 파일 내에 한글이 있는데, 이를 사용하기 위해 encoding='cp949' 필요 
table = pd.read_csv('TempDir/SeoulTemp.csv', encoding='cp949')
```

```python
# pandas 라이브러리를 이용해, 데이터의 통계 요약 결과 보기
# 원래 데이터(table)에서 '날짜', '지점'의 열을 제외한 일부 데이터를 추출
table_partial = table.drop(['날짜','지점'], axis=1)

# 추출한 데이터를 대상으로 요약 결과 출력
print(table_partial.describe())
```

```python
dates = table['날짜']
num_data = len(dates)
print('데이터 개수 : ', num_data)
```

```python
# x 축의 값을 생성하기
date_min, date_max = 1, 31
x_values = list(range(date_min,date_max+1))
print(x_values)
```

```python
plt.rc('font', family='Malgun Gothic')  # 맑은 고딕 글꼴 사용하기
plt.plot(x_values, table['평균기온'])
plt.xlabel('5월')
plt.ylabel('기온(섭씨)')
plt.title('2021년 5월 평균기온 변화')
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')  # 맑은 고딕 글꼴 사용하기
# 그래프 색상 지정하기 : k(black), b(blue), r(red)
plt.plot(x_values, table['평균기온'], 'k',
         x_values, table['최저기온'], 'b',
         x_values, table['최고기온'], 'r')
plt.xlabel('5월')
plt.ylabel('기온(섭씨)')
plt.title('2021년 5월 기온 변화')
plt.legend(['평균', '최처', '최고'])  # 범례 설정
plt.show()
```

```python
# 기상청 그래프와 비교하기위해 기상청 그래프를 화면에 출력
import matplotlib.image as img

image = img.imread('TempDir/기상청-2021.05-기온그래프.png')  # 파일읽기
plt.imshow(image)  # 이미지 불러오기
plt.axis('off')  # x,y축 표시하지 않기
plt.show()
```

```python
# 최고 기온이 최고였던 날짜는?
highest_temp_data = np.array(table['최고기온'])

print('최고기온 중 최고기온: ', np.max(highest_temp_data))
print('최고기온 중 최고기온 날짜: 5월', np.argmax(highest_temp_data), '일')
print('')  # 출력값을 구분하기 위해, 공백을 출력

# 첫번째 데이터가 1일부터 시작하므로, argmax 결과를 그대로 날짜로 사용할 수 있음
print('최고기온 중 최저기온: ', np.min(highest_temp_data))
print('최고기온 중 최저기온 날짜: 5월', np.argmin(highest_temp_data), '일')
print('')  # 출력값을 구분하기 위해, 공백을 출력

# 소수점 둘째자리 까지만 출력하기 위해 round(값, 2) 를 사용
print('최고기온의 평균: ', round(np.mean(highest_temp_data), 2))
```

```python
# 10도 단위로 구분하여 히스토그램 그리기
# 즉, 0~10도, 10~20도, 20~30도, 30~40도에 해당하는 기온 발생 횟수에 대한 히스토그램 그리기
plt.hist(highest_temp_data, bins=[0, 10, 20, 30, 40])
plt.xticks([0,10,20,30,40])  # x 축에 원하는 값만 표시되도록 하기
plt.show()
```

```python
# 5도 단위로 구분하여 히스토그램 그리기
# 즉, 0~5도, 5~10도, ..., 35~40도에 해당하는 기온 발생 횟수에 대한 히스토그램 그리기
plt.hist(highest_temp_data, bins=[0,5,10,15,20,25,30,35,40])
plt.xticks([0,5,10,15,20,25,30,35,40])  # x 축에 원하는 값만 표시되도록 하기
plt.show()
```
